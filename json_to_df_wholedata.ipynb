```python
import os
import pandas as pd
```

```python
data_dir = '/home/jh20/Data/pet/반려견_샘플데이터_B_221209/라벨링 데이터'
json_list =  os.listdir(data_dir)
json_list[0]
```

```python
import json

with open(os.path.join(data_dir, json_list[1]), 'r') as f:
    json_data = json.load(f)
#print(json.dumps(json_data))
```

```python
print(json.dumps(json_data, indent = '\t'))
```

```python
from collections import defaultdict
meta_dict = defaultdict(list)
```

```python
json_data['metadata'].keys()
```

```python
json_key_list = []
for i in json_data['metadata'].keys():
    if i == 'type':
        json_key_list.append('type')
    
    else:
        json_key_list = json_key_list + list(json_data['metadata'][i].keys())

#json_key_list
meta_dict = {i:[] for i in json_key_list}
meta_dict['diseaseName'] = []
meta_dict
```

```python
for i in json_list:
    with open(os.path.join(data_dir, i), 'r', encoding="UTF-8") as f:
        json_data = json.load(f)
    
    # diseaseName 있는 파일과 없는 파일 구분하려고 굳이 추가
    disease_list = list()
    for col in json_data['metadata'].keys():
        if col == 'type':
            disease_list.append('type')
        else:
            disease_list = disease_list + list(json_data['metadata'][col].keys())
    
    # diseaseName 없는 파일은
    if 'diseaseName' not in disease_list:
        meta_dict['diseaseName'].append('nan') # diaseaseName 값으로 'nan' append
        #
        for col in json_data['metadata'].keys():
            if col == 'type':
                meta_dict[col].append(json_data['metadata'][col])
            else:
                for col2 in json_data['metadata'][col].keys():
                    meta_dict[col2].append(json_data['metadata'][col][col2])
    
    # diseaseName 있는 파일
    else:
        for col in json_data['metadata'].keys():
            if col == 'type':
                meta_dict[col].append(json_data['metadata'][col])
            else:
                for col2 in json_data['metadata'][col].keys():
                    meta_dict[col2].append(json_data['metadata'][col][col2])
```
